```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# 
def apply_smoothing(image):
    return cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)

def detect_edges(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.medianBlur(gray, 7)
    edges = cv2.Canny(blurred, threshold1=50, threshold2=100)
    edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    return edges

def apply_cartoon_effect(image):
    smooth = apply_smoothing(image)
    edges = detect_edges(image)
    cartoon = cv2.addWeighted(smooth, 0.8, edges, 0.2, 0)
    return cartoon

#  
images = []
image_names = ["behzad.jpg", "samapic.png"]  #      

for image_name in image_names:
    if os.path.exists(image_name):
        img = cv2.imread(image_name)
        if img is not None:
            images.append(img)
        else:
            print(f"   : {image_name}")
    else:
        print(f"  '{image_name}'  !")

#     
cartoon_images = []
for img in images:
    cartoon_img = apply_cartoon_effect(img)
    cartoon_images.append(cartoon_img)

#   (  )
for i in range(len(images)):
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(' ')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(cartoon_images[i], cv2.COLOR_BGR2RGB))
    plt.title(' ')
    plt.axis('off')
    plt.show()
```
